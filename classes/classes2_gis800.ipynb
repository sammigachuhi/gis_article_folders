Let's create a class to display some geometries.

````python
# Modules to be used in creating the class objects
import pandas as pd
import geopandas as gpd
````

Our class will be responsible for creating a dataframe of lon-lat coordinates of a place. The goal is to create a class that creates a geodataframe from the lon-lat coordinates and finally plots a map using the coordinates. In classes, it's best to start simple and progress along. That should be your lesson 101 in `classes` for python!

````python
# Modules to be used in creating the class objects
import pandas as pd
import geopandas as gpd

class Place():
    """Describe our place and show it"""
    def __init__(self, names,  longitudes, latitudes):
        # Create variables for the names, longitudes and latitudes
        self.names = names
        self.longitudes = longitudes
        self.latitudes = latitudes

    def dataTable(self):
        # Create a dataframe from the above variables of names, longitudes, and latitudes
        self.dataTable = pd.DataFrame({
            'Name': [self.names],
            'Longitude': [self.longitudes], 
            'Latitude': [self.latitudes]})
        
    def showDataFrame(self):
        # Print the dataframe
        print(self.dataTable)
````

## Inheritance

As if classes don't look complicated enough, we shall introduce a concept known as **inheritance**. Inheritance is simply a code picking up the characteristics of a different class, albeit to perform a similar or totally different function. Consider that you are a Parent. I know, some of us are yet to be :). Your child will obviously inherit some of your qualities, perhaps eyes, voice, complexion teeth and personality (maybe!). Now in the physique of your child, the body already has some components of your qualities within his/her DNA. The same applies to Object Oriented Programming.

For our case, we shall create a class called `GeoSpatial` that inherits all the properties of the `Place` class, but uses these properties to create a geodataframe. The `Place` class simply stops at the creation of a dataframe. The `GeoSpatial` class will go beyond this. Consider the `Place` class as the parent, and the `Geospatial` class as the child. 

````python
# Modules to be used in creating the class objects
import pandas as pd
import geopandas as gpd

class Place():
    """Describe our place and show it"""
    def __init__(self, names,  longitudes, latitudes): # Everything in the brackets is an argument. Self is always the first argument in a class method defintion
        self.names = names
        self.longitudes = longitudes
        self.latitudes = latitudes

    def dataTable(self):
        self.dataTable = pd.DataFrame({
            'Name': [self.names],
            'Longitude': [self.longitudes], 
            'Latitude': [self.latitudes]})
        
    def showDataFrame(self):
        print(self.dataTable)
    
class GeoSpatial(Place):
    # This class is to take the arguments from Place class and use them to create a geodataframe
    def __init__(self, names, longitudes, latitudes, crs):
        super().__init__(names, longitudes, latitudes)
        self.crs = crs
        print("Name: ", self.names, "\nLongitudes: ", self.longitudes, "\nLatitudes: ", self.latitudes, "\nCRS: ", crs)
            
    def geoDataFrame(self):
        # Create a geodataframe
        self.geoDataFrame = gpd.GeoDataFrame(self.names, 
                         geometry=gpd.points_from_xy(self.longitudes, self.latitudes), 
                         crs=crs)
        print(self.geoDataFrame)  
        
````

Look at the `GeoSpatial` class a little closer. We have introduced the `def __init__()` function to initialize the same attributes as those of `Place` class. We have introduced another function, the `super()` function. The purpose of this latter function is to inherit the same instance attributes used in `Place` class to `GeoSpatial` class.

````python
# Create an object place1 for the Geospatial class
place1 = GeoSpatial("vasco", 40.1276701, -3.2236304, "EPSG:4326")
````

````python
# Create a second object for Geospatial class --place2
place2 = GeoSpatial("gedi", 39.9623916, -3.334307, 'EPSG:4326')
````

This class has been excitingly too long. We have already seen our `GeoSpatial` class can draw a webmap. Time for recess? Almost. But first let's stretch the lesson a bit further. Can classes inherit from another subclass, such as from `Geospatial` class as in this case? Yes. The inheritance is limitless, much like your child can beget a grandchild. We shall make this final subclass do the work of saving our webmap to the directory. 

````python
# Modules to be used in creating the class objects
import pandas as pd
import geopandas as gpd

# Classes using contextily
import contextily as cx
import matplotlib.pyplot as plt

class Place():
    """Describe our place and show it"""
    def __init__(self, names,  longitudes, latitudes): # Everything in the brackets is an argument. Self is always the first argument in a class method defintion
        self.names = names
        self.longitudes = longitudes
        self.latitudes = latitudes

    def dataTable(self):
        # Create a dataframe of names, longitudes and latitudes
        self.dataTable = pd.DataFrame({
            'Name': self.names,
            'Longitude': self.longitudes, 
            'Latitude': self.latitudes})
        
    def showDataFrame(self):
        # Prints out a neat dataframe created from dataTable above
        print(self.dataTable)
    
class GeoSpatial(Place):
    def __init__(self, names, longitudes, latitudes, crs):
        # This class is to take the arguments from Place class and use them to create a geodataframe
        super().__init__(names, longitudes, latitudes)
        self.crs = crs
        print("Name: ", self.names, "\nLongitudes: ", self.longitudes, "\nLatitudes: ", self.latitudes, "\nCRS: ", crs)
            
    def geoDataFrame(self):
        # Creates geodataframe
        self.geoDataFrame = gpd.GeoDataFrame(self.names, 
                         geometry=gpd.points_from_xy(self.longitudes, self.latitudes), 
                         crs= self.crs)
        
    def showGeoDataFrame(self):
        # Prints geodataframe
        print(self.geoDataFrame)
        
    def webMap(self):
        # Plots webmap
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        # Here contextily plots the webmap
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
class Map(GeoSpatial):
    def __init__(self, names, longitudes, latitudes, crs):
        super().__init__(names, longitudes, latitudes, crs)
        
    def saveMap(self, directory, dpi, orientation):
        # Saves webmap to directory
        super().geoDataFrame()
        self.directory = directory
        self.dpi = dpi
        self.orientation = orientation
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
        plt.savefig(fname=self.directory, dpi=self.dpi, orientation=self.orientation)
````

Once again the `def __init__()` has been used to initialize the attributes that will go into this class. We have also used a less known trick of `super()` function. That is to [access the methods of its parent class](https://www.codingem.com/super-in-python/?utm_source=pocket_reader). In our case, for the `saveMap` method to work, it needs some of the properties of `geoDataFrame` method from the `GeoSpatial` class. It also needs the variables of `directory`, `dpi` and `orientation` for our file path, pixel resolution and relative positioning of the map respectively. The `savefig` function enables our `saveMap` method save the map into our computer's directory.

The `facecolor` parameter has been assigned a default argument `0` because it is optional anywhere, and missing out the argument will result in an error. If one would like to specify a value for the `facecolor` parameter, be it a string or integer, they can do so and the `saveMap` method will happily execute their wish.

````python
# Modules to be used in creating the class objects
import pandas as pd
import geopandas as gpd

# Classes using contextily
import contextily as cx
import matplotlib.pyplot as plt

class Place():
    """Describe our place and show it"""
    def __init__(self, names,  longitudes, latitudes): # Everything in the brackets is an argument. Self is always the first argument in a class method defintion
        self.names = names
        self.longitudes = longitudes
        self.latitudes = latitudes

    def dataTable(self):
        # Create a dataframe of names, longitudes and latitudes
        self.dataTable = pd.DataFrame({
            'Name': self.names,
            'Longitude': self.longitudes, 
            'Latitude': self.latitudes})
        
    def showDataFrame(self):
        # Prints out a neat dataframe created from dataTable above
        print(self.dataTable)
    
class GeoSpatial(Place):
    def __init__(self, names, longitudes, latitudes, crs):
        # This class is to take the arguments from Place class and use them to create a geodataframe
        super().__init__(names, longitudes, latitudes)
        self.crs = crs
        print("Name: ", self.names, "\nLongitudes: ", self.longitudes, "\nLatitudes: ", self.latitudes, "\nCRS: ", crs)
            
    def geoDataFrame(self):
        # Creates geodataframe
        self.geoDataFrame = gpd.GeoDataFrame(self.names, 
                         geometry=gpd.points_from_xy(self.longitudes, self.latitudes), 
                         crs= self.crs)
        
    def showGeoDataFrame(self):
        # Prints geodataframe
        print(self.geoDataFrame)
        
    def webMap(self):
        # Plots webmap
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        # Here contextily plots the webmap
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
class Map(GeoSpatial):
    def __init__(self, names, longitudes, latitudes, crs):
        super().__init__(names, longitudes, latitudes, crs)
        
    def saveMap(self, directory, dpi, orientation, facecolor='auto'): # We've set facecolor to auto to make it optional. If one want's specification, they can change the value
        super().geoDataFrame()
        self.directory = directory
        self.dpi = dpi
        self.orientation = orientation
        self.facecolor = facecolor
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
        plt.savefig(fname=self.directory, dpi=self.dpi, orientation=self.orientation, facecolor=self.facecolor)
````
