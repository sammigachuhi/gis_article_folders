# Introduction
Classes are our biggest nightmare. However, next to `for` loops and `functions`, `classes` are the next critical component of python you've got to learn. But mastering them can be definitely challenging. In python, everything is an object (to confirm). This includes variables, functions, lists, tuples, dictionaries and et cetera (https://www.javatpoint.com/what-is-an-object-in-python). To make this concept of consider the following very simple example:

````python
max = 24
````

The variable `max` is an object. Let's take it as though this is a real life person called 'Max' and his age is 24. Thus the `max` variable, which is by extension also an *object* is an integer --because it is assigned to the value 24. Having seen what a class is, let's introduce what's the relationship between clases and objects. Objects are an *instance* of a class. That is, the `class` acts as a blue-print of what the object will look like, and function. Consider the below example of our former example of the `max` object.

````python
# Get class of the object -max
type(max)
````

We have used the `type` function to investigate the class type of our `max` object. You can find out the class of an object in python using the `type` function. Because almost everything in python is considered an object, this make it to be referred to as an **Object Oriented Programming** language. 

In python, we create the blueprint of our object instances using the `class` keyword. Let's first create a simple examble that will resonate with everyone (https://pynative.com/python-classes-and-objects/).

````python
class Friend():
    """This is a class to show simple information about a student."""
    def __init__(self, name, sex, age):
        # Attributes of your friend
        self.name = name
        self.sex = sex
        self.age = age 
        
    def describe(self):
        # Describe to a person your friend's characteristics
        print("My best friend is called ", self.name, " a ", self.sex, "and is ", self.age, " years.")
````

What did we just do in the above example? In python, classes are creating using the below format. Remember classes act as blueprints, have that in mind as you look at the outline below.

```python
class class_name:
    '''This is a docstring. I have created a new class'''
    def __init__(self, x, y, z)
    <statement 1>
    <statement 2>
    .
    .
    <statement N>
```

The `def __init__(self, ...)` is a keyword/method used to create the instance of an object. It is for this reason referred to as a constructor. The argument `self` is a naming convention referring to the current object. Finally, last but not least the statement in docstrings (""" """) is used to describe what the class does. 

To close all this, objects are created using the following convention:

```python
<object-name> = <class-name>(<arguments>)  
```

Let's try creating an object our `Friend` class using the above convention. Our friend is called 'Diana'.


````python
# Create class of your first friend
friend1 = Friend('Diana', 'Female', '22')
````

To describe our friend, we have to use the method `describe(self)` which we had created within the `Friend` class.

````python
friend1.describe()
````

Got it? Let's finally make our class look better by tweaking a few things and cross over to spatial matters.

````python
class Friend():
    """This class describes the name, sex and age of my friend."""
    def __init__(self, name, sex, age):
        # Attributes of your friend
        self.name = name
        self.sex = sex
        self.age = age 
        
    def describe(self):
        # Describe to a person your friend's characteristics
        print("My best friend is: ", "\nName: ", self.name, "\nSex: ", self.sex, "\nAge: ", self.age)
        
friend2 = Friend('Jonathan', 'Male', 25)
friend2.describe()
````

````python
# Classes using contextily
import contextily as cx
import matplotlib.pyplot as plt

class Place():
    """Describe our place and show it"""
    def __init__(self, names,  longitudes, latitudes):
        self.names = names
        self.longitudes = longitudes
        self.latitudes = latitudes

    def dataTable(self):
        self.dataTable = pd.DataFrame({
            'Name': self.names,
            'Longitude': self.longitudes, 
            'Latitude': self.latitudes})
        
    def showDataframe(self):
        print(self.dataTable)
    
class GeoSpatial(Place):
    def __init__(self, names, longitudes, latitudes, crs):
        super().__init__(names, longitudes, latitudes)
        self.crs = crs
        print("Name: ", self.names, "\nLongitudes: ", self.longitudes, "\nLatitudes: ", self.latitudes, "\nCRS: ", crs)
            
    def geoDataFrame(self):
        self.geoDataFrame = gpd.GeoDataFrame(self.names, 
                         geometry=gpd.points_from_xy(self.longitudes, self.latitudes), 
                         crs= self.crs)
        
    def showGeoDataFrame(self):
        print(self.geoDataFrame)
        
    def webMap(self):
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        # for name, lon, lat in zip(self.names, self.longitudes, self.latitudes):
            # self.geoDataFrame.annotate(name, xy=(lon, lat), xytext=(-4, 2), textcoords='offset points')
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
class Map(GeoSpatial):
    def __init__(self, names, longitudes, latitudes, crs, directory, dpi, orientation):
        super().__init__(names, longitudes, latitudes, crs)
        self.directory = directory
        self.dpi = dpi
        self.orientation = orientation
        
    def saveMap(self):
        super().geoDataFrame()
        fig, ax = plt.subplots(figsize=(8, 6))
        webMap = self.geoDataFrame.plot(ax=ax)
        cx.add_basemap(ax, crs=self.geoDataFrame.crs, source=cx.providers.OpenStreetMap.Mapnik)
        
        plt.savefig(self.directory)
        
````
